# Example of imgwpipe package usage


Importing hydro_daily module, pandas and seaborn with matplotlib for data visualisation.

```python
import hydro_daily as hd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading data for desired time range and hydrological station.
In this case it will be between 2015 and 2017 hydrological years.

```python
data = hd.getrange(2015, 2017, stationid=150190310)
```

```python
data.head()
```

Using seaborn and matplotlib to plot the change of the river flow over time.

```python
sns.set_theme(style="darkgrid", rc = {'figure.figsize':(16,4)}, font_scale=1.4)
sns.lineplot(x="date", y="flow", data=data).set(title='Flow of the Rudawa river')
plt.xlabel('Date')
plt.ylabel('River flow [$m^3/s$]')
plt.show()
```
